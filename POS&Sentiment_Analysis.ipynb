```python
corpus= []
for i in range(0,len(messages)):
  review = re.sub('[^a-zA-Z]',' ',messages['text'][i])
  review = review.lower()
  review = review.split()

  review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
  review = ' '.join(review)
  corpus.append(review)
```

TF-IDF(Term Frequency - Inverse Document Frequency)


* S1-->good boy
* S2-->good girl
*  S3-->boy girl good



*   TF = No.of repetition of words in sentence/No. of words in sentence
*   IDF = Loge(No. of sentences/ No. of sentences containing that word)



ADV


*   Word importance captured.
*   Fixed size input.

Disadvantage

*   Sparsity exist Out of vocabulary







TF-IDF

```python
import pandas as pd
message = pd.read_csv('sms.csv',index_col = 0)
```

```python
messages
```

```python
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
corpus = []
for i in range(0,len(messages)):
  review = re.sub('[^a-zA-Z]',' ',messages['text'][i])
  review = review.lower()
  review = review.split()
  review = [lemmatizer.lemmatize(word) for word in review if not word in stopwords.words('english')]
  review = ' '.join(review)
  corpus.append(review)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=100)
X = tfidf.fit_transform(corpus).toarray()
```
